## Daten eines RGB-Bildes laden (Rohdaten ohne Header)

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Daten aus Datei einlesen (byteweise --> unit8)
img = np.fromfile("rgb.data", dtype='uint8')
# Daten aus Array in Bildform bringen 3 Byte pro Pixel einer Matrix 640x480
img_r=img.reshape((480,640,3))
# Pyplot mit Bild verbinden
plt.imshow(img_r)
# Bild anzeigen
plt.show()
```

## Bildaten von BGR in RGB umwandel

```python
import numpy as np
# import cv2
import matplotlib.pyplot as plt


# Daten aus Datei einlesen (byteweise --> unit8)
img = np.fromfile("rgb.data", dtype='uint8')

# Daten aus Array in Bildform bringen 3 Byte pro Pixel einer Matrix 640x480
img_r = img.reshape((480,640,3))

# Daten konvertieren mit CV2 Funktion
img_k = cv2.cvtColor(img_r, cv2.COLOR_BGR2RGB)

# Struktur der Bildmatrix ausgeben
print(img_k.shape)

# Pyplot mit Bild verbinden
plt.imshow(img_k)

# Bild anzeigen
plt.show()
```

## Einzelne Farbkanäle aus Bilddaten extrahieren

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Daten aus Datei einlesen (byteweise --> unit8)
img = np.fromfile("rgb.data", dtype='uint8')
# Daten aus Array in Bildform bringen 3 Byte pro Pixel einer Matrix 640x480
img_r = img.reshape((480,640,3))
# Anzeige Pixel (0,0)
print("RGB "+str(img_r[0][0]))
# Extraktion Grünkanal
img_r = img_r[ :, :, 1:2]
# Anzeige Pixel (0,0)
print("Grün "+str(img_r[0][0]))
# Vervielfältigen auf wieder 3 Werte pro Pixel
img_k = cv2.cvtColor(img_r, cv2.COLOR_GRAY2RGB)
# Verbinden der Bildmatrix mit Pyplot
plt.imshow(img_k)
# Anzeigen der Bildmatrix
plt.show()
```

## Bilddaten aufhellen/abdunkeln

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Daten aus Datei einlesen (byteweise --> unit8)
img = np.fromfile("rgb.data", dtype='uint8')
# Daten aus Array in Bildform bringen 3 Byte pro Pixel einer Matrix 640x480
img_r=img.reshape((480,640,3))
for i in range(img_r.shape[0]):
    for j in range(img_r.shape[1]):
        for k in range(img_r.shape[2]):
            img_r[i][j][k]=img_r[i][j][k] + 50
# Pyplot mit Bild verbinden
plt.imshow(img_r)
# Bild anzeigen
plt.show()
```
